```python
# Import required packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the dataset
dataset = pd.read_csv('emp1.csv')
print(dataset)
```

```python
X = dataset.iloc[:, 0:1].values  # Features (independent variables)
y = dataset.iloc[:, 1].values    # Target (dependent variable)
```

```python
# Display shapes to verify the split
print("\nShape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
```

```python
# Create scalers for both X and y
scaler_X = StandardScaler()
scaler_y = StandardScaler()
```

```python
# Fit and transform X
X = scaler_X.fit_transform(X)
```

```python
# Fit and transform y (need to reshape as it's 1D)
y = scaler_y.fit_transform(y.reshape(-1, 1))
```

```python
# Visualize data
plt.scatter(X, y, color='blue')
plt.title('Data Visualization')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Train Support Vector Regressor model
from sklearn.svm import SVR

# Create SVR model with default parameters
sv_regressor = SVR()

# Fit the model to the training data
sv_regressor.fit(X, y)
```

```python
# Visualize Support Vector Regression result
plt.scatter(X, y, color='blue')  # Original data points
plt.plot(X, sv_regressor.predict(X), color='green')  # Prediction line
plt.title('Support Vector Regressor')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#do prediction
y_pred = sv_regressor.predict([[7.5]])
print(y_pred)
y_pred = scaler_y.inverse_transform(y_pred.reshape(-1, 1))  # Reshape to 2D array
print(y_pred)
```
